```python
import pandas as pd
import db
```

```python
def plot_artist(artist, year):
    sql = """
        SELECT Y.edition, S.score
        FROM (
            SELECT DISTINCT edition FROM Occurance
        ) Y
        LEFT JOIN ( 
            SELECT O.edition, SUM(2000 - position) AS score
            FROM Occurance O
            INNER JOIN Song S ON S.rowid = O.song
            INNER JOIN Artist A ON A.rowid = S.artist
            WHERE name LIKE '%%%s%%'
            GROUP BY O.edition
        ) S ON S.edition = Y.edition
        ORDER BY Y.edition
    """ % artist
    df = db.dataframe(sql)
    df = df.fillna(0)
    df = df.set_index('edition')
    ax = df.plot.bar()
    ax.set_title(artist)
    ax.get_legend().remove()
    ax.get_children()[year-1999].set_color('brown')
```

```python
# 2022
dead_artists = [ "Faithless", "Specials", "Olivia Newton John", "Fleetwood Mac", "Doe maar", "Foo Fighters", "Jerry Lee Lewis"]

for a in dead_artists:
    plot_artist(a, 2022)
```
