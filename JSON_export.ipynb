```python
import pandas as pd
import db
import json
```

```python
sql = """
SELECT O.edition, O.position, O.artist, O.title, O.year 
FROM Occurance O
WHERE position <= 100
AND edition >= 2016
ORDER By O.edition, O.position
"""

df = db.dataframe(sql)
df.describe()
```

```python
edition = 0
output = []
current_edition = {}
for row in db.raw_sql(sql):
    # print(row)
    if (row[0] != edition):
        edition = row[0]
        current_edition = {
            "edition": row[0],
            "songs": []
        }
        output.append(current_edition)
    current_edition["songs"].append({
        "position" : row[1],
        "artist"   : row[2],
        "title"    : row[3],
        "year"     : row[4]
    })

json_object = json.dumps(output, indent = 2) 
# print(json_object)
with open("sample.json", "w") as outfile:
    outfile.write(json_object)
```
